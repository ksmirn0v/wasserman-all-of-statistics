# Expectation

### Expectation of Discrete Random Variables

**Notation**

$\mathcal{X} = X(\Omega)$

---

**Definition**

The expectation of a discrete random variabl $X$ with PMF $p$ is

$\mathbb{E}(X) = \sum_{x\in\mathcal{X}} xp(x)$, when this sum is "well-defined" (otherwise, it doesn't exist)

---

**Definition**

Let $a_1, a_2, ... \in \mathbb{R}: \sum_{i=1}^\infty a_i = \sum_{i: a_i\geq 0} a_i + \sum_{i: a_i < 0} a_i = b + c$

$\sum_{i=1}^\infty a_i$ is well-defined if either $b$ or $c$ is finite.

### Expectation of Continuous Random Variables

**Definition**

The expectation of a random variable $X$ with density $f(x)$ is

$\mathbb{E}(X) = \int_{-\infty}^\infty x f(x) dx$, when this integral is "well-defined" (otherwise, it doesn't exist)

---

**Definition**

Let $a = \int_{-\infty}^0 xf(x)dx$ and $b = \int_0^\infty xf(x)dx$

$\int_{-\infty}^\infty x f_X(x) dx$ is well-defined if either $a$ or $b$ is finite.

### Expectation Rule

**Definition**

$g(X)$ is a random variable if $X$ is a random variable and $g:\mathbb{R}\to\mathbb{R}$ is a measurable function.

---

**Theorem**

If $X$ is a random variable and $g:\mathbb{R}\to\mathbb{R}$ is a measurable function, then:

- Discrete case

$\mathbb{E}(g(X)) = \sum_{x\in\mathcal{X}} g(x)p(x)$, when the sum is "well-defined"

- Continuous case

$\mathbb{E}(g(X)) = \int_{-\infty}^\infty g(x)f(x)dx$, when the integral is "well-defined"

### Properties of Expectation

**Theorem**

Suppose $X$ and $Y$ are random variables, such that $\mathbb{E}(|X|) < \infty$ and $\mathbb{E}(|Y|) < \infty$:

- $\forall a\in\mathbb{R}, \mathbb{E}(a) = a$ ($\mathbb{E}(X) = a$ if $\forall\omega\in\Omega, X(\omega) = a$)

- $\forall a\in\mathbb{R}, \mathbb{E}(aX) = a\mathbb{E}(X)$

- $\mathbb{E}(X+Y) = \mathbb{E}(X) + \mathbb{E}(Y)$

- if $X\geq 0$, then $\mathbb{E}(X)\geq0$

- if $X\leq Y$, then $\mathbb{E}(X)\leq \mathbb{E}(Y)$

- $\mathbb{E}(I_A(X)) = \mathbb{P}(X\in A)$

---

**Theorem**

If $X_1, X_2, ...$ is a sequence of random variables, such that $\forall i, X_i\geq 0$, then:

- $\mathbb{E}(\sum_{i=1}^\infty X_i) = \sum_{i=1}^\infty\mathbb{E}(X_i)$

### Mean, Variance, and Moments

Let $X$ be a random variable

**Definition**

The mean of $X$ is $\mu(X) = \mathbb{E}(X)$

---

**Definition**

The variance of $X$ is $\sigma^2(X) = \mathbb{E}((X-\mu)^2)$

---

**Definition**

The standard deviation of $X$ is $\sigma(X) = \sqrt{\sigma^2(X)}$

---

**Definition**

The $k^{th}$ moment of $X$ is $m_k(X) = \mathbb{E}(X^k)$

---

**Definition**

The $k^{th}$ central moment of $X$ is $c_k(X) = \mathbb{E}((X-\mu)^k)$

---

**Theorem**

The variance can be defined as:

$\sigma^2(X) = \mathbb{E}(X^2) - (\mathbb{E}(X))^2$

### Multivariate Random Variables

**Definition**

$g(X)$ is a random vector if $X\in\mathbb{R}^d$ is a random vector and $g:\mathbb{R}^d\to\mathbb{R}^k$ is a measurable function.

---

**Definition**

- Discrete case

$\mathbb{E}(g(X)) = \sum_{x\in\mathcal{X}}g(x)p(x)$, when the sum is "well-defined"

- Continuous

$\mathbb{E}(g(X)) = \int_{\mathbb{R}^d} g(x)f(x)dx$, when the integral is "well-defined"

---

**Definition**

The conditional expectation of $X$ given $Y=y$ is defined by:

- Discrete case

$\mathbb{E}(X|Y=y) = \sum_{x\in\mathcal{X}}xp(x|y)$, when the sum is "well-defined"

- Continuous case

$\mathbb{E}(X|Y=y) = \int_{-\infty}^\infty xf(x|y)dx$, whe the integral is "well-defined"

---

**Definition**

The covariance of $X$ and $Y$ is:

$Cov(X,Y) = \mathbb{E}((X-\mu(X))(Y-\mu(Y))$

- $Cov(X,X) = \sigma^2(X)$

- $Cov(X,Y) = \mathbb{E}(XY) - \mathbb{E}(X)\mathbb{E}(Y)$

- $X,Y$ are independent $\Rightarrow Cov(X,Y) = 0$

---

**Definition**

The Pearson correlation coefficient of $X$ and $Y$:

$\rho(X,Y) = \frac{Cov(X,Y)}{\sigma(X)\sigma(Y)}$
