# Measure Theory

### Banach-Tarski paradox

### Sigma Algebra

Given a set $\Omega$, a $\sigma$-algebra on $\Omega$ is a collection  $\mathcal{A}\subset 2^\Omega$:
- $\mathcal{A}\neq\emptyset$
- if $E\in\mathcal{A}$, then $E^c\in\mathcal{A}$ (closed under complement)
- if $E_1, E_2, ...\in\mathcal{A}$, then $\cup_{i=1}^\infty\in\mathcal{A}$ (closed under countable union)


Example:  
$\Omega = \{0,1\} \Rightarrow 2^\Omega = \{\emptyset, \{0\}, \{1\}, \{0,1\}\}$

---
---

Prove that $\Omega\in\mathcal{A}$.

If $E\in\mathcal{A} \Rightarrow E^c\in\mathcal{A} \Rightarrow E\cup E^c = \Omega\in\mathcal{A}$

---

Prove that $\emptyset\in\mathcal{A}$.

If $\Omega\in\mathcal{A} \Rightarrow \Omega^c\in\mathcal{A} \Rightarrow \Omega^c=\emptyset\in\mathcal{A}$

---

Prove that $\mathcal{A}$ is closed under countable intersections.

Suppose $E_1, E_2,...\in\mathcal{A}$.

$\cap_{i=1}^\infty E_i = \cap_{i=1}^\infty (E_i^c)^c = (\cup_{i=1}^\infty E_i^c)^c\in\mathcal{A}$

---
---

Given $C\subset 2^\Omega$, the $\sigma$-algebra, generated by $C$ (_i.e._ $\sigma(C)$), is the smallest $\sigma$-algebra, containing $C$ $\rightarrow$ $\sigma(C) = \cap_{\mathcal{A}_i \supset C} \mathcal{A}_i$

$\sigma(C)$ always exists because:
- $2^\Omega$ is a $\sigma$-algebra
- intersection of $\sigma$-algebras is a $\sigma$-algebra

If $\Omega=\mathbb{R}$, then Borel $\sigma$-algebra $\mathcal{B} = \sigma(\tau)$, where $\tau = \{\text{open sets of $\mathbb{R}$}\}$

### Measure

A measure $\mu$ on $\Omega$ with $\sigma$-algebra $\mathcal{A}$ is a function $\mu:\mathcal{A}\rightarrow[0,\infty]$:
- $\mu(\emptyset) = 0$
- $\mu(\cup_{i=1}^\infty E_i) = \sum_{i=1}^\infty \mu(E_i)$ for any $E_1, E_2,...\in\mathcal{A}$ where $E_1,E_2,...$ are pairwise disjoint sets

### Probability measure

A probability measure $\mathbb{P}$ on $(\Omega,\mathcal{A})$ is a function $\mathbb{P}:\mathcal{A}\to[0,1]$:
- $\mathbb{P}(\Omega) = 1$

Example - finite set - uniform distribution

$\Omega = \{1,...,n\}, \mathcal{A} = 2^\Omega, \forall k\in\Omega: \mathbb{P}(\{k\}) = \frac{1}{n}$

$\mathbb{P}(\{1,2,4\}) = \mathbb{P}(\{1\}\cup\{2\}\cup\{4\}) = \mathbb{P}(\{1\}) + \mathbb{P}(\{2\}) + \mathbb{P}(\{4\}) = \frac{3}{n}$

Example - countable infinite set - geometric distribution

$\Omega=\{1,2,3,...\}, \mathcal{A}=2^\Omega, \forall k\in\Omega: \mathbb{P}(\{k\}) = p(1-p)^{k-1}\text{, where }p - \text{ probability of a single outcome}$

Example - uncountable set - exponential distribution

$\Omega=[0,\infty), \mathcal{A}=\mathcal{B}([0,\infty)), \forall x > 0:\mathbb{P}([0,x))=1-e^{-x}, \forall x \geq 0:\mathbb{P}(\{x\})=0$

### Lebesgue measure

$\Omega = \mathbb{R}, \mathcal{A} = \mathcal{B}(\mathbb{R}), \forall a,b\in\mathbb{R}, a< b:\mu((a,b)) = b-a$

### Measure properties

Let $(\Omega, \mathcal{A}, \mu)$ be a measure space.
- monotonicity: if $E,F\in\mathcal{A}$ and $E\subset F$, then $\mu(E)\leq\mu(F)$
- sub-additivity: if $E_1, E_2,...\in\mathcal{A}$, then $\mu(\cup_{i=1}^\infty E_i)\leq\sum_{i=1}^\infty \mu(E_i)$
- continuity from below: if $E_1, E_2,...\in\mathcal{A}$ and $E_1\subset E_2\subset...$, then $\mu(\cup_{i=1}^\infty E_i) = \lim_{i\to\infty}\mu (E_i)$
- continuity from above: if $E_1, E_2,...\in\mathcal{A}$ and $E_1\supset E_2\supset...$ and $\mu (E_1) < \infty$, then $\mu(\cap_{i=1}^\infty E_i) = \lim_{i\to\infty}\mu (E_i)$

### Probability properties

Let $(\Omega,\mathcal{A},\mathbb{P})$ be a probability measure space with $E, F, E_i\in\mathcal{A}$
- $\mathbb{P}(E\cup F) = \mathbb{P}(E) + \mathbb{P}(F)$ if $E\cap F=\emptyset$
- $\mathbb{P}(E\cup F) = \mathbb{P}(E) + \mathbb{P}(F) - \mathbb{P}(E\cap F)$
- $\mathbb{P}(E) = 1 - \mathbb{P}(E^c)$
- $\mathbb{P}(E\cap F^c) = \mathbb{P}(E) - \mathbb{P}(E\cap F)$
- $\mathbb{P}(\cup_{i=1}^n E_i) = \sum_{i=1}^n \mathbb{P}(E_i) - \sum_{i<j}^n \mathbb{P}(E_i\cap E_j) + \sum_{i<j<k}^n \mathbb{P}(E_i\cap E_j \cap E_k) - ... (-1)^{n+1} \mathbb{P}(E_1\cap E_2\cap ... \cap E_n)$
- $\mathbb{P}(\cup_{i=1}^n E_i) \leq \sum_{i=1}^n \mathbb{P}(E_i)$ and $\mathbb{P}(\cup_{i=1}^\infty E_i) \leq \sum_{i=1}^\infty \mathbb{P}(E_i)$

### Boreal probability measures

Let $\Omega = \mathbb{R}, \mathcal{B}(\mathbb{R})$.

A Borel measure on $\mathbb{R}$ is a measure on $(\mathbb{R}, \mathcal{B}(\mathbb{R}))$.

A Cumulatve Distribution Function (CDF) is a function $F: \mathbb{R}\to\mathbb{R}$:
- $F$ is non-decreasing ($x\leq y\Rightarrow F(x)\leq F(y)$)
- $F$ is right continuous ($\lim_{x\downarrow a}F(x) = F(a)$)
- $\lim_{x\to\infty} F(x) = 1$
- $\lim_{x\to-\infty} F(x) = 0$

---

Theorem

if $F$ is a CDF, then there's a unique Borel probability measure on $\mathbb{R}$:
- $\forall x\in\mathbb{R}:\mathbb{P}((-\infty, x]) = F(x)$

if $\mathbb{P}$ is a Borel probability measure on $\mathbb{R}$, then there's a unique CDF $F$:
- $\forall x\in\mathbb{R}: F(x) = \mathbb{P}((-\infty,x])$

That is, there's an equivalence between CDFs and Borel probability measures.

# Conditional probability and independence

Let $\Omega$ - sample space

Let $(\Omega, \mathcal{A})$ be a measure space

Let "event" = measurable set = a set in $\mathcal{A}$

### Conditional probability

If $\mathbb{P}(B) > 0$, then $\mathbb{P}(A|B) = \frac{\mathbb{P}(A\cap B)}{\mathbb{P}(B)}$

If $\mathbb{P}(B) > 0$, then $\mathbb{Q}(A) = \mathbb{P}(A|B)$ defines a probability measure $\mathbb{Q}$

### Independence

1) Events $A$ and $B$ are independent if $\mathbb{P}(A\cap B) = \mathbb{P}(A)\mathbb{P}(B)$

2) Events $A_1, A_2,...,A_n$ are mutually independent if $\forall S\subset\{1,...,n\}: \mathbb{P}(\cap_{i\in S} A_i) = \prod_{i\in S} \mathbb{P}(A_i)$

3) Events $A_1, A_2, ...$ are mutually independent if $\forall S\subset\{1,2,...\}: \mathbb{P}(\cap_{i\in S} A_i) = \prod_{i\in S} \mathbb{P}(A_i)$

4) Events $A$ and $B$ are conditionally independent given $C$ if $\mathbb{P}(A\cap B|C) = \mathbb{P}(A|C)\mathbb{P}(B|C)$

5) Events $A_1, A_2,...,A_n$ are conditionally independent given $C$ if $\forall S\subset\{1,2,...,n\}: \mathbb{P}(\cap_{i\in S} A_i |C) = \prod_{i\in S} \mathbb{P}(A_i|C)$

6) Events $A_1, A_2,...$ are conditionally independent given $C$ if $\forall S\subset\{1,2,...\}: \mathbb{P}(\cap_{i\in S} A_i |C) = \prod_{i\in S} \mathbb{P}(A_i|C)$

---

**Theorem**

Suppose $P(B) > 0$. $A$ and $B$ are independent $\iff$ $\mathbb{P}(A|B) = \mathbb{P}(A)$

---

**Theorem - Bayes' Rule**

$\mathbb{P}(B|A) = \frac{\mathbb{P}(A|B)\mathbb{P}(B)}{\mathbb{P}(A)}$ where $\mathbb{P}(A) > 0$ and $\mathbb{P}(B) > 0$

---

**Theorem - Chain Rule**

if $A_1, A_2,...,A_n$ and $\mathbb{P}(A_1\cap A_2 \cap ... \cap A_n) > 0$ then
$\mathbb{P}(A_1\cap A_2 \cap ... \cap A_n) = \mathbb{P}(A_1)\mathbb{P}(A_2|A_1)\mathbb{P}(A_3|A_1\cap A_2)...\mathbb{P}(A_n|A_1\cap A_2 \cap ... \cap A_{n-1})$

---

A partition of $\Omega$ is a finite/countable collection of sets $B_1, B_2,... \subset\Omega$:
- $\cup_{i=1} = \Omega$
- $B_i\cap B_j =\emptyset$ where $i\neq j$

**Theorem - Partition Rule**

$\mathbb{P}(A) = \sum_{i=1}\mathbb{P}(A\cap B_i)$ for any partition $B_1, B_2,...$ of $\Omega$
