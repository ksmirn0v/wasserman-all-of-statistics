# Measure Theory

### Banach-Tarski paradox

### Sigma Algebra

Given a set $\Omega$, a $\sigma$-algebra on $\Omega$ is a collection  $\mathcal{A}\subset 2^\Omega$:
- $\mathcal{A}\neq\emptyset$
- if $E\in\mathcal{A}$, then $E^c\in\mathcal{A}$ (closed under complement)
- if $E_1, E_2, ...\in\mathcal{A}$, then $\cup_{i=1}^\infty\in\mathcal{A}$ (closed under countable union)


Example:  
$\Omega = \{0,1\} \Rightarrow 2^\Omega = \{\emptyset, \{0\}, \{1\}, \{0,1\}\}$

---
---

Prove that $\Omega\in\mathcal{A}$.

If $E\in\mathcal{A} \Rightarrow E^c\in\mathcal{A} \Rightarrow E\cup E^c = \Omega\in\mathcal{A}$

---

Prove that $\emptyset\in\mathcal{A}$.

If $\Omega\in\mathcal{A} \Rightarrow \Omega^c\in\mathcal{A} \Rightarrow \Omega^c=\emptyset\in\mathcal{A}$

---

Prove that $\mathcal{A}$ is closed under countable intersections.

Suppose $E_1, E_2,...\in\mathcal{A}$.

$\cap_{i=1}^\infty E_i = \cap_{i=1}^\infty (E_i^c)^c = (\cup_{i=1}^\infty E_i^c)^c\in\mathcal{A}$

---
---

Given $C\subset 2^\Omega$, the $\sigma$-algebra, generated by $C$ (_i.e._ $\sigma(C)$), is the smallest $\sigma$-algebra, containing $C$ $\rightarrow$ $\sigma(C) = \cap_{\mathcal{A}_i \supset C} \mathcal{A}_i$

$\sigma(C)$ always exists because:
- $2^\Omega$ is a $\sigma$-algebra
- intersection of $\sigma$-algebras is a $\sigma$-algebra

If $\Omega=\mathbb{R}$, then Borel $\sigma$-algebra $\mathscr{B} = \sigma(\tau)$, where $\tau = \{\text{open sets of $\mathbb{R}$}\}$

### Measure

A measure $\mu$ on $\Omega$ with $\sigma$-algebra $\mathcal{A}$ is a function $\mu:\mathcal{A}\rightarrow[0,\infty]$:
- $\mu(\emptyset) = 0$
- $\mu(\cup_{i=1}^\infty E_i) = \sum_{i=1}^\infty \mu(E_i)$ for any $E_1, E_2,...\in\mathcal{A}$ where $E_1,E_2,...$ are pairwise disjoint sets

### Probability measure

A probability measure is a measure $\mathbb{P}$:
- $\mathbb{P}(\Omega) = 1$
